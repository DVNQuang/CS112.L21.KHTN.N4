```
Function Fibonacci(n):
IF n = 1 or n = 2 THEN RETURN 0
ELSE 
  DECLARE a, a1, a2
  FOR i = 3 to i = n
      a = a1 + a2
      a2 = a1
      a1 = a
  RETURN a
```

````

def fibonacci(n):
  if (n < 2):
    return n
  else:
    return fibonacci(n - 1) + fibonacci(n - 2)2
    

def solution(n, k):
  return n * fibonacci(2 * k + 1)

n = int(input())
k = int(input())
rs = solution(n, k)
print(rs)
````
